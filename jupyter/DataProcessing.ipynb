```python
from scipy.ndimage import gaussian_filter
import numpy as np

def process_data(filename, outfile, processedfile, sigma=1.):
    with open(filename, "r") as f:
        f.readline()
        lines = f.readlines()
        f.close()
    lis = []
    for line in lines:
        d = line.rstrip().split('\t')
        lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
    lis
    with open(outfile, "w") as g:
        for j in range(len(lis[0])):
            g.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        g.close()
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] = gaussian_filter(lis[i][1:], sigma=sigma)
    with open(processedfile, "w") as h:
        for j in range(len(lis[0])):
            h.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        h.close()

def thresholded_process_data(filename, outfile, processedfile, sigma=1.):
    with open(filename, "r") as f:
        f.readline()
        lines = f.readlines()
        f.close()
    lis = []
    dic = {}
    for line in lines:
        d = line.rstrip().split('\t')
        dic[d[0]] = list(map(float, d[1:]))
        if (d[0] == "#OTU ID"):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
        elif (np.mean(dic[d[0]]) > 300):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
    with open(outfile, "w") as g:
        for j in range(len(lis[0])):
            g.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        g.close()
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] = gaussian_filter(lis[i][1:], sigma=sigma)
    with open(processedfile, "w") as h:
        for j in range(len(lis[0])):
            h.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        h.close()

def thresholded_process_data2(filename, outfile, processedfile, sigma=1.):
    with open(filename, "r") as f:
        f.readline()
        lines = f.readlines()
        f.close()
    lis = []
    dic = {}
    for line in lines:
        d = line.rstrip().split('\t')
        dic[d[0]] = list(map(float, d[1:]))
        if (d[0] == "#OTU ID"):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
        elif (np.mean(dic[d[0]]) > 100):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
    with open(outfile, "w") as g:
        for j in range(len(lis[0])):
            g.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        g.close()
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] = gaussian_filter(lis[i][1:], sigma=sigma)
    with open(processedfile, "w") as h:
        for j in range(len(lis[0])):
            h.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        h.close()
        
def thresholded_normalized_process_data(filename, outfile, processedfile, sigma=1.):
    with open(filename, "r") as f:
        f.readline()
        lines = f.readlines()
        f.close()
    lis = []
    dic = {}
    for line in lines:
        d = line.rstrip().split('\t')
        dic[d[0]] = list(map(float, d[1:]))
        maxval = max(dic[d[0]])
        if (d[0] == "#OTU ID"):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
        elif (np.mean(dic[d[0]]) > 300):
            lis.append([d[0]] + [float(d[i])/maxval for i in range(1,len(d))])
    with open(outfile, "w") as g:
        for j in range(len(lis[0])):
            g.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        g.close()
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] = gaussian_filter(lis[i][1:], sigma=sigma)
    with open(processedfile, "w") as h:
        for j in range(len(lis[0])):
            h.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        h.close()
        
def thresholded_normalized_process_data2(filename, outfile, processedfile, sigma=1.):
    with open(filename, "r") as f:
        f.readline()
        lines = f.readlines()
        f.close()
    lis = []
    dic = {}
    for line in lines:
        d = line.rstrip().split('\t')
        dic[d[0]] = list(map(float, d[1:]))
        if (d[0] == "#OTU ID"):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
        elif (np.mean(dic[d[0]]) > 300):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
    with open(outfile, "w") as g:
        for j in range(len(lis[0])):
            g.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        g.close()
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] = gaussian_filter(lis[i][1:], sigma=sigma)
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] /= max(lis[i][1:])
    with open(processedfile, "w") as h:
        for j in range(len(lis[0])):
            h.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        h.close()
        
def thresholded_normalized_process_data3(filename, outfile, processedfile, sigma=1.):
    with open(filename, "r") as f:
        f.readline()
        lines = f.readlines()
        f.close()
    lis = []
    dic = {}
    for line in lines:
        d = line.rstrip().split('\t')
        dic[d[0]] = list(map(float, d[1:]))
        if (d[0] == "#OTU ID"):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
        elif (np.mean(dic[d[0]]) > 10):
            lis.append([d[0]] + [float(d[i]) for i in range(1,len(d))])
    with open(outfile, "w") as g:
        for j in range(len(lis[0])):
            g.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        g.close()
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] = gaussian_filter(lis[i][1:], sigma=sigma)
    for i in range(len(lis)):
        if lis[i][0] != '#OTU ID':
            lis[i][1:] /= max(lis[i][1:])
    with open(processedfile, "w") as h:
        for j in range(len(lis[0])):
            h.write("\t".join(str(lis[i][j]) for i in range(len(lis))) + "\n")
        h.close()
```
