```r
plot.microbiome <- function(file){
    d <- read.delim(header=T, file)
    temp <- melt(d, id="X.OTU.ID", measure=c("k__Bacteria.p__Actinobacteria","k__Bacteria.p__Bacteroidetes", "k__Bacteria.p__Cyanobacteria", "k__Bacteria.p__Firmicutes", "k__Bacteria.p__Fusobacteria", "k__Bacteria.p__Proteobacteria"))
    return(ggplot(temp, aes(x=X.OTU.ID, y=value, colour=variable, group=variable)) + geom_line())
}
```
